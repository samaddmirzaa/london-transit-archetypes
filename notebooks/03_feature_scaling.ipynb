#### Feature Scaling

##### Importing Libraries and Loading Data

```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import RobustScaler
import joblib

project_root = Path.cwd().parent
processed_dir = project_root /'data'/'processed'
models_dir = project_root /'models'

features = pd.read_parquet(processed_dir/'station_features.parquet')
```

##### Robust Scaling

```python
# Robust Scaling works best as it handles outliers better

scaler = RobustScaler()
scaled_values = scaler.fit_transform(features)
```

```python
features_scaled = pd.DataFrame(
    scaled_values,
    index=features.index,
    columns=features.columns
)
```

```python
# Saving

features_scaled.to_parquet(processed_dir/'station_features_scaled.parquet')
joblib.dump(scaler, models_dir/'scaler.joblib')
```
